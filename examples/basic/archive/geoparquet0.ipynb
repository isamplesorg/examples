### CesiumJS Visualization (Experimental)
This section provides a lightweight CesiumJS equivalent to the earlier Lonboard point visualization.

Features:
- Loads a subset of points (configurable) directly from the same GeoParquet file via DuckDB
- Builds an inbrowser GeoJSON object and renders it with CesiumJS
- Enables clientside clustering for scalability
- No Cesium Ion token required (uses OpenStreetMap imagery); you can add one later for premium terrain/imagery

Performance notes:
- Rendering thousands of point billboards/entities is fine; >25k may get sluggish. Use clustering (enabled) or downsample.
- For very large datasets, consider: tiling (3D Tiles), serverside aggregation, or heatmaps.

Next steps you could try later:
- Replace simple points with color by category (material/context)
- Add popups with richer HTML
- Stream chunks instead of embedding all JSON inline
- Switch to `GeoJsonDataSource.clustering` tuning (pixelRange / minimumClusterSize)


```python
import duckdb, json, os, math, textwrap
from IPython.display import HTML, display

# --- Config ---
PARQUET_PATH = '/Users/raymondyee/Data/iSample/2025_02_20_10_30_49/isamples_export_2025_02_20_10_30_49_geo.parquet'
MAX_POINTS = 10000          # Upper bound of points to embed (tune for performance)
SAMPLE_FRACTION = None       # e.g., 0.01 for 1% random sample; overrides MAX_POINTS if set
RANDOM_SEED = 42

# --- Extract (lon, lat, id, label) subset via DuckDB ---
con = duckdb.connect()
query_base = f"""
    SELECT 
        sample_location_longitude AS lon,
        sample_location_latitude  AS lat,
        sample_identifier         AS sample_id,
        COALESCE(label, sample_identifier) AS label
    FROM read_parquet('{PARQUET_PATH}')
    WHERE sample_location_longitude IS NOT NULL
      AND sample_location_latitude  IS NOT NULL
"""

if SAMPLE_FRACTION is not None:
    points_df = con.sql(query_base + f" USING SAMPLE {SAMPLE_FRACTION * 100}% (bernoulli, {RANDOM_SEED})").df()
else:
    points_df = con.sql(query_base + f" LIMIT {MAX_POINTS}").df()

point_count = len(points_df)
print(f"Prepared {point_count} points for CesiumJS (SAMPLE_FRACTION={SAMPLE_FRACTION}, MAX_POINTS={MAX_POINTS})")

# --- Convert to GeoJSON FeatureCollection ---
features = []
for row in points_df.itertuples(index=False):
    lon, lat, sample_id, label = row
    if math.isnan(lon) or math.isnan(lat):
        continue
    features.append({
        "type": "Feature",
        "geometry": {"type": "Point", "coordinates": [float(lon), float(lat)]},
        "properties": {"sample_id": sample_id, "label": label}
    })
geojson_obj = {"type": "FeatureCollection", "features": features}
geojson_str = json.dumps(geojson_obj)  # embed directly (small subset)

# --- HTML/JS Template for Cesium ---
html = f"""
<div style='height: 650px; border:1px solid #ccc;'>
  <div id='cesiumContainer' style='width:100%; height:100%;'></div>
</div>
<link rel='stylesheet' href='https://unpkg.com/cesium/Build/Cesium/Widgets/widgets.css'>
<script src='https://unpkg.com/cesium/Build/Cesium/Cesium.js'></script>
<script>
  // Embedded GeoJSON data
  const geojsonData = {geojson_str};

  // Basic Cesium setup
  window.CESIUM_BASE_URL = 'https://unpkg.com/cesium/Build/Cesium/';
  const viewer = new Cesium.Viewer('cesiumContainer', {{
    animation: false,
    timeline: false,
    geocoder: false,
    homeButton: true,
    sceneModePicker: true,
    baseLayerPicker: true,
    navigationHelpButton: false,
    infoBox: true,
    selectionIndicator: true,
    fullscreenButton: false,
    shadows: false,
    terrainProvider: Cesium.EllipsoidTerrainProvider(),
    imageryProvider: new Cesium.OpenStreetMapImageryProvider({{ url: 'https://a.tile.openstreetmap.org/' }})
  }});

  // Load GeoJSON
  Cesium.GeoJsonDataSource.load(geojsonData, {{
    clampToGround: false
  }}).then(ds => {{
      viewer.dataSources.add(ds);
      viewer.flyTo(ds.entities, {{ duration: 0 }});

      // Enable clustering for performance
      const cl = ds.clustering;
      cl.enabled = true;
      cl.pixelRange = 25;            // expand cluster radius (px)
      cl.minimumClusterSize = 5;     // show clusters until fewer than this

      // Optional styling callback for clusters
      function customStyle() {{
        const removeListeners = [];
        cl.clusterEvent.addEventListener((clusteredEntities, cluster) => {{
          cluster.label.show = true;
          cluster.label.text = clusteredEntities.length.toLocaleString();
          cluster.label.fillColor = Cesium.Color.WHITE;
          cluster.label.outlineColor = Cesium.Color.BLACK;
          cluster.label.outlineWidth = 2;
          cluster.billboard.show = true;
          cluster.billboard.id = cluster.label.id = clusteredEntities[0];
          cluster.billboard.image = 'data:image/svg+xml;base64,' + btoa(`<?xml version='1.0'?><svg xmlns='http://www.w3.org/2000/svg' width='64' height='64'><circle cx='32' cy='32' r='28' fill='#d33' stroke='white' stroke-width='4'/></svg>`);
        }});
        removeListeners.push(() => cl.clusterEvent.removeEventListener());
      }}
      customStyle();

      // Style individual (unclustered) points
      for (const e of ds.entities.values) {{
        if (e.point) {{
          e.point.pixelSize = 6;
          e.point.color = Cesium.Color.RED.withAlpha(0.7);
          e.point.outlineColor = Cesium.Color.WHITE;
          e.point.outlineWidth = 1.0;
        }} else if (e.billboard) {{
          // fallback if future conversion yields billboards
          e.billboard.scale = 0.6;
        }}
        // Attach a description (HTML) for infoBox
        const props = e.properties || {{}};
        const sid = props.sample_id ? props.sample_id.getValue() : 'N/A';
        const label = props.label ? props.label.getValue() : sid;
        e.description = `<table class='cesium-infoBox-defaultTable'><tr><th>Sample ID</th><td>${{sid}}</td></tr><tr><th>Label</th><td>${{label}}</td></tr></table>`;
        e.name = label;
      }}
  }}).catch(err => {{
    console.error('Cesium GeoJSON load error:', err);
  }});
</script>
"""

# Display
display(HTML(html))
```

### CesiumJS Troubleshooting & Alternate Embed
If the earlier Cesium cell (inline `<script>` tags) did not render a globe, it's likely because JupyterLab blocks execution of script tags inside raw HTML outputs for security. This alternate approach:

- Separately displays a container `<div>`
- Dynamically injects the Cesium script via a JavaScript execution cell (which JupyterLab allows)
- Rebuilds the GeoJSON subset (selfcontained)
- Should work in both classic Notebook and JupyterLab

If this still fails:
1. Open the browser dev tools Console for errors about network/CORS or missing symbols.
2. Verify network access to `unpkg.com`.
3. Try reducing `MAX_POINTS`.
4. If using a very restrictive network, download Cesium locally and serve from a local path.

The next code cell implements this alternate method.

```python
import duckdb, json, math
from IPython.display import display, HTML, Javascript

# --- Config (alternate embed) ---
PARQUET_PATH = '/Users/raymondyee/Data/iSample/2025_02_20_10_30_49/isamples_export_2025_02_20_10_30_49_geo.parquet'
MAX_POINTS = 8000

con = duckdb.connect()
points_df = con.sql(f"""
    SELECT 
        sample_location_longitude AS lon,
        sample_location_latitude  AS lat,
        sample_identifier         AS sample_id,
        COALESCE(label, sample_identifier) AS label
    FROM read_parquet('{PARQUET_PATH}')
    WHERE sample_location_longitude IS NOT NULL
      AND sample_location_latitude  IS NOT NULL
    LIMIT {MAX_POINTS}
""").df()

features = []
for row in points_df.itertuples(index=False):
    lon, lat, sid, label = row
    if math.isnan(lon) or math.isnan(lat):
        continue
    features.append({
        'type': 'Feature',
        'geometry': {'type': 'Point', 'coordinates': [float(lon), float(lat)]},
        'properties': {'sample_id': sid, 'label': label}
    })
geojson_obj = {'type': 'FeatureCollection', 'features': features}
geojson_json = json.dumps(geojson_obj)

container_id = 'cesium_alt_container'

# Create container first (script tags inside HTML here won't run, so keep it pure markup)
html_container = f"""
<div id='{container_id}' style='width:100%;height:650px;border:1px solid #ccc'>
  <div style='padding:8px;font-family:system-ui, sans-serif;font-size:13px;background:#f5f5f5;'>
    CesiumJS (dynamic injection). Points: {len(features):,}
  </div>
</div>
"""

display(HTML(html_container))

# JavaScript loader & viewer init executed explicitly so JupyterLab allows it
js_code = f"""
(async function() {{
  const containerId = '{container_id}';
  const geojsonData = {geojson_json};

  function loadScript(src) {{
    return new Promise((resolve, reject) => {{
      if (document.querySelector(`script[src="${{src}}"]`)) return resolve();
      const s = document.createElement('script');
      s.src = src; s.async = true; s.onload = resolve; s.onerror = reject; document.head.appendChild(s);
    }});
  }}
  function loadCss(href) {{
    if (document.querySelector(`link[href="${{href}}"]`)) return;
    const l = document.createElement('link'); l.rel='stylesheet'; l.href = href; document.head.appendChild(l);
  }}

  loadCss('https://unpkg.com/cesium/Build/Cesium/Widgets/widgets.css');
  await loadScript('https://unpkg.com/cesium/Build/Cesium/Cesium.js');

  if (!window.Cesium) {{
    console.error('Cesium failed to load');
    return;
  }}
  window.CESIUM_BASE_URL = 'https://unpkg.com/cesium/Build/Cesium/';

  const host = document.getElementById(containerId);
  // Remove any prior globe (re-run safe)
  const oldViewerDiv = host.querySelector('.cesium-viewer');
  if (oldViewerDiv) oldViewerDiv.remove();
  const globeDiv = document.createElement('div');
  globeDiv.style.cssText = 'width:100%;height:600px;';
  host.appendChild(globeDiv);

  const viewer = new Cesium.Viewer(globeDiv, {{
    animation: false,
    timeline: false,
    geocoder: false,
    homeButton: true,
    sceneModePicker: true,
    baseLayerPicker: true,
    navigationHelpButton: false,
    fullscreenButton: false,
    selectionIndicator: true,
    infoBox: true,
    imageryProvider: new Cesium.OpenStreetMapImageryProvider({{ url: 'https://a.tile.openstreetmap.org/' }})
  }});

  // Expose viewer on DOM so later cells can restyle it easily
  host.__viewer__ = viewer;
  globeDiv.__viewer__ = viewer;
  const cesiumViewerEl = globeDiv.querySelector('.cesium-viewer');
  if (cesiumViewerEl) cesiumViewerEl.__viewer__ = viewer;

  const ds = await Cesium.GeoJsonDataSource.load(geojsonData, {{ clampToGround: false }});
  viewer.dataSources.add(ds);
  viewer.flyTo(ds.entities, {{ duration: 0 }});

  // Also expose datasource for later style tweaks
  host.__datasource__ = ds;

  const cl = ds.clustering;
  cl.enabled = true;
  cl.pixelRange = 25;
  cl.minimumClusterSize = 5;

  cl.clusterEvent.addEventListener((clusteredEntities, cluster) => {{
    cluster.label.show = true;
    cluster.label.text = clusteredEntities.length.toLocaleString();
    cluster.label.fillColor = Cesium.Color.WHITE;
    cluster.label.outlineColor = Cesium.Color.BLACK;
    cluster.label.outlineWidth = 2;
    cluster.billboard.show = true;
    cluster.billboard.image = 'data:image/svg+xml;base64,' + btoa(`<?xml version='1.0'?><svg xmlns='http://www.w3.org/2000/svg' width='64' height='64'><circle cx='32' cy='32' r='28' fill='#1976d2' stroke='white' stroke-width='4'/></svg>`);
  }});

  // Individual points styling
  for (const e of ds.entities.values) {{
    if (e.point) {{
      e.point.pixelSize = 6;
      e.point.color = Cesium.Color.fromBytes(200,50,50,200);
      e.point.outlineColor = Cesium.Color.WHITE;
      e.point.outlineWidth = 1;
    }}
    const props = e.properties || {{}};
    const sid = props.sample_id ? props.sample_id.getValue() : 'N/A';
    const label = props.label ? props.label.getValue() : sid;
    e.description = `<b>Sample ID:</b> ${{sid}}<br><b>Label:</b> ${{label}}`;
    e.name = label;
  }}

  console.log('Cesium alternate embed complete. Entities:', ds.entities.values.length);
}})();
"""

display(Javascript(js_code))
```

### CesiumJS style tweaks: smaller clusters and points
Seeing large blue cluster bubbles is expected with clustering on. The next cell:
- Shrinks cluster circles and scales them by count
- Uses smaller unclustered points
- Lets you tune clustering (pixel range, minimum size) and point size quickly

Tip: If you want no clusters at all, set CLUSTER_ENABLED = False in the next cell and re-run it.

```python
# Re-render Cesium with smaller clusters/points
from IPython.display import Javascript

CLUSTER_ENABLED = True
CLUSTER_PIXEL_RANGE = 18        # radius in pixels for clustering neighborhood
CLUSTER_MIN_SIZE = 10           # minimum entities to form a cluster
POINT_PIXEL_SIZE = 3            # unclustered point size

# The previous alternate renderer used container id 'cesium_alt_container'
js_style = f"""
(function() {{
  const host = document.getElementById('cesium_alt_container');
  if (!host) {{ console.warn('Cesium container not found'); return; }}
  const viewerEl = host.querySelector('.cesium-viewer');
  if (!viewerEl) {{ console.warn('No existing Cesium viewer found'); return; }}

  // Try to reference the Cesium Viewer instance bound to this container
  // We stored viewer only locally before; if not accessible, re-style via datasource traversal
  const canvases = host.querySelectorAll('.cesium-viewer');
  // Best-effort: grab last viewer in host and infer DataSourceDisplay from window.Cesium
  const CesiumRef = window.Cesium;
  if (!CesiumRef) {{ console.warn('Cesium not on window'); return; }}

  // Walk all data sources and restyle
  const viewers = Array.from(document.querySelectorAll('.cesium-viewer')).map(v => v.__viewer__).filter(Boolean);
  // Fallback: Without private handle, detect via global registry (not standard). We'll instead attempt to restyle via first viewer in page.
  let viewer = (viewers && viewers[viewers.length-1]) || null;
  if (!viewer) {{
    // Heuristic: try to find the Cesium widget attached to the container
    try {{
      // eslint-disable-next-line no-underscore-dangle
      const widget = CesiumRef.Widget && CesiumRef.Widget._instances && CesiumRef.Widget._instances.find(w => host.contains(w.container));
      viewer = widget && widget.viewer;
    }} catch (e) {{}}
  }}

  if (!viewer) {{ console.warn('Could not resolve viewer instance; re-run the alternate Cesium cell first.'); return; }}

  viewer.dataSources._dataSources.forEach(ds => {{
    if (CLUSTER_ENABLED) {{
      ds.clustering.enabled = true;
      ds.clustering.pixelRange = {CLUSTER_PIXEL_RANGE};
      ds.clustering.minimumClusterSize = {CLUSTER_MIN_SIZE};
    }} else {{
      ds.clustering.enabled = false;
    }}

    // Reset cluster event and apply lighter, smaller cluster bubbles
    ds.clustering.clusterEvent._listeners = [];
    ds.clustering.clusterEvent.addEventListener((clusteredEntities, cluster) => {{
      const n = clusteredEntities.length;
      const radius = Math.max(10, Math.min(24, 10 + Math.log2(n+1)*4)); // scale gently with count
      cluster.label.show = true;
      cluster.label.text = n.toLocaleString();
      cluster.label.fillColor = CesiumRef.Color.WHITE;
      cluster.label.outlineColor = CesiumRef.Color.BLACK;
      cluster.label.outlineWidth = 2;
      cluster.billboard.show = true;
      const svg = `<?xml version='1.0'?><svg xmlns='http://www.w3.org/2000/svg' width='64' height='64'><circle cx='32' cy='32' r='${{radius}}' fill='#1e88e5' stroke='white' stroke-width='3'/></svg>`;
      cluster.billboard.image = 'data:image/svg+xml;base64,' + btoa(svg);
      cluster.billboard.scale = 0.6; // make base 64 icon smaller overall
    }});

    // Style unclustered points smaller
    ds.entities.values.forEach(e => {{
      if (e.point) {{
        e.point.pixelSize = {POINT_PIXEL_SIZE};
        e.point.color = CesiumRef.Color.fromBytes(210,50,50,200);
        e.point.outlineColor = CesiumRef.Color.WHITE;
        e.point.outlineWidth = 1;
      }}
    }});
  }});

  console.log('Applied style tweaks to Cesium entities');
}})();
"""

display(Javascript(js_style))
```
